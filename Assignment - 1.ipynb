```python
#importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

```python
#reading data
data = pd.read_csv('DA1-Q1.csv')
y = np.array(data["Height"]).reshape(-1, 1)
x = np.array(data["Weight"]).reshape(-1, 1)
```

```python
#initial scatter plot
plt.scatter(x = x, y = y)
plt.show()
```

```python
#split data to test and train
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state = 1)
```

```python
#fit the model
model = LinearRegression()
model.fit(x_train, y_train)
```

```python
#scoring the model
print(model.score(x_train, y_train))
print(model.score(x_test, y_test))
```

```python
#exploring data
y_pred = model.predict(x_test)
print(x_test)
print(y_pred)
```

```python
#plotting regression line
plt.scatter(x_test, y_test, alpha = 0.5)
plt.plot(x_test, y_pred, color = "red")
plt.show()
```

```python
# error calculation
print('Mean Squared Error: ', mean_squared_error(y_test, y_pred))
```

```python
#importing libraries
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

```python
#reading data
df = pd.read_csv('DA1-Q2.csv')

y = np.array(df['price']).reshape(-1, 1)
x = np.array(df[['bedrooms', 'bathrooms', 'sqft_living', 'sqft_lot', 'floors', 'waterfront', 'view', 'condition', 'sqft_above', 'sqft_basement', 'yr_built', 'yr_renovated']])
```

```python
#split data to test and train
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state = 1, train_size = 0.65, test_size = 0.35)
```

```python
#fit the model
model = LinearRegression()
model.fit(x_train, y_train)
```

```python
#scoring the model
print(model.score(x_train, y_train))
print(model.score(x_test, y_test))
```

```python
#exploring data
y_pred = model.predict(x_test)
print(x_test)
print(y_pred)
```

```python
#plotting

fig = plt.figure()
ax = fig.gca(projection ='3d')
 
ax.scatter(y_test, y_pred)
ax.view_init(45, 45)

plt.xlabel('Actual')
plt.ylabel('Predicted')

plt.show()
```

```python
#error calculation
print('Mean Squared Error: ', mean_squared_error(y_test, y_pred))
```
